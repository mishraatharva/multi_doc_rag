```python
# from langchain.document_loaders import TextLoader
from langchain_community.document_loaders import TextLoader
```

```python
loader = TextLoader(r"U:\LLMops\data\data.txt", encoding="utf8")
documents = loader.load()
```

```python
documents[0].page_content[:500]  # Print the first 500 characters of the first documen
```

```python
# from langchai.text_splitter import RecursiveCharacterTextSplitter # old
from langchain_text_splitters import RecursiveCharacterTextSplitter # new
```

```python
text_splitter=RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=20)
```

```python
text_chunks=text_splitter.split_documents(documents)
```

```python
text_chunks
```

```python
import faiss
print(faiss.__version__)
```

```python
# from langchain.embeddings import OpenAIEmbeddings # old
# from langchain_openai import OpenAIEmbeddings # new
fireworks_api_key = os.environ["FIREWORKS_API_KEY"]

# from langchain.vectorstores import FAISS # old
from langchain_community.vectorstores import FAISS # new
```

```python
# embeddings=OpenAIEmbeddings()

from langchain_fireworks import FireworksEmbeddings

embeddings_model = FireworksEmbeddings(
    model="nomic-ai/nomic-embed-text-v1.5",
)

vector_db = FAISS.from_documents(text_chunks, embeddings_model)
```

```python
# vectorstore=FAISS.from_documents(text_chunks, embeddings)
```

```python
vector_db
```

```python
retriever=vector_db.as_retriever()
```

```python
# Perform similarity search
query = "What is the Key Characteristics of Agentic AI?"
docs = vector_db.similarity_search(query, k=4)

# Display the results
for i, doc in enumerate(docs):
    print(f"Document {i+1}:")
    print(doc.page_content)
    print("-" * 50)
```

```python
from langchain_core.prompts import ChatPromptTemplate

template="""You are an assistant for question-answering tasks.
Use the following pieces of retrieved context to answer the question.
If you don't know the answer, just say that you don't know.
Use ten sentences maximum and keep the answer concise.
Question: {question}
Context: {context}
Answer:
"""
```

```python
prompt=ChatPromptTemplate.from_template(template)
```

```python
prompt
```

```python
# from langchain.schema.output_parser import StrOutputParser #old
from langchain_core.output_parsers import StrOutputParser #new

# from langchain_community.output_parsers import StrOutputParser
```

```python
output_parser=StrOutputParser()
```

```python
# from langchain.chat_models import ChatOpenAI
from langchain_groq import ChatGroq

llm_model=ChatGroq(model_name="openai/gpt-oss-120b")
```

```python
# from langchain.schema.runnable import RunnablePassthrough
from langchain_core.runnables import RunnablePassthrough


rag_chain = (
    {"context": retriever,  "question": RunnablePassthrough()}
    | prompt
    | llm_model
    | output_parser
)
```

```python
rag_chain.invoke("tell me about Agentic AI")
```
